```python
# 1번
lt=[]
result_lt=[]
str1='c:K:g:E'
for kk in str1:
  lt.append(kk)
for pp in lt:
  if(pp != ':'):
    result_lt.append(pp)
    result_str1 = ''.join(result_lt)
lt2=sorted(result_str1.lower())
result_str1 = ''.join(lt2) 
print(result_str1)
```

```python
# 3번
import numpy as np
num = list(np.random.randint(1,10000,10))
max = num[0]
for kk in num:
  for i in range(len(num)):
    if(max< num[i]):
      max = num[i]
  if kk == max:
    print('오른쪽 값이 최댓값입니다.',end=' ')
    print(max, end=' ')
  else:
    print(kk,end=' ')
```

```python
# 4번
result_sum = 0 
num = str(input())
result_num = num.split(' ');
for kk in result_num:
  result = int(kk)*int(kk)
  result_sum = int(result)+int(result_sum)
print(result_sum%10)
```
